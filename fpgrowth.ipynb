```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv("I:/NMK Files/collage/New folder/Apriori_Algorithm_French_Store_Dataset_1709027182336 (1).csv")
df.head()
```

```python
# to transform data
te = TransactionEncoder()
te1 = te.fit(df).transform(df)
data = pd.DataFrame(te1)
data
```

```python
frequent_itemset = fpgrowth(data,min_support=0.001)
frequent_itemset 
```

```python
res = association_rules(frequent_itemset,metric = "confidence",min_threshold=1)
res
```

```python
res2 = res[["lift","antecedents"]]
res2
```
