```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\Selva Karthik\Downloads\Mall_Customers_1712655402567.csv")
df
```

```python
data = df.iloc[:,3:5].values
data
```

```python
import scipy.cluster.hierarchy as sch

dend = sch.dendrogram(sch.linkage(data,method = "complete"))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 5,metric = 'euclidean',linkage="single")
labels_ = cluster.fit_predict(data)
df['Cluster'] = labels_
labels_
```

```python

plt.scatter(data[:,0],data[:,1],c=labels_)
```

```python
print(df[df['Cluster']==1].describe())
print(df[df['Cluster']==2].describe())
print(df[df['Cluster']==3].describe())
print(df[df['Cluster']==4].describe())
```
